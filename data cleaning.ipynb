```python
## New task
#Working with country and drug names\

data = pd.read_csv("/data/notebook_files/Drug data.csv")
data.head(10)
```

```python
unique_countries = data["Country"].unique()
unique_countries
```

```python
df_list = []
for i in unique_countries:
  df_list.append(data[data["Country"] == i])

print(df_list)
```

```python
max_df_list = []
for i in df_list:
  max_df_list.append(i.max())

final = pd.DataFrame(max_df_list)
final.head(20)
```

```python
temp = {
    "A": [1,2,3,4,5],
    "B": [6,None,4,3,0],
    "C": [10,30,None,20,11]

}

temp = pd.DataFrame(temp)
temp.head()
```

```python
temp.max()
```

```python
temp["B"].fillna(temp["B"].max(),inplace=True)
```

```python
temp
```

```python
temp["C"].fillna(temp["C"].median(),inplace=True)
```

```python
temp
```

```python
temp.transpose()
```

```python
temp.transform(func='sqrt')
```

```python
df = temp.transform(func = ['sqrt',lambda x: x**5])
```
